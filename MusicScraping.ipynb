```python
import requests 
import re
import pandas as pd
from bs4 import BeautifulSoup
key = 'C'
response = requests.get("https://www.flutetunes.com/tunes.php?key={}".format(key)) 
html_source = response.text   
import urllib2
import os.path
import time
import numpy as np
```

```python
soup = BeautifulSoup(html_source, 'html.parser')
search_count = soup.find('div', id="content")
ilist = []
i = 0
rows = search_count.findAll('tr')
for row in rows:
    a = row.findAll('a')
    ilist.append(re.findall("id=(\d+)", str(a)))
    i += 1

ilist = filter(None, ilist)
keylist = [int(x.pop(0)) for x in ilist]

print len(keylist), keylist
```

```python
df = pd.DataFrame(index = keylist, columns = ['Title', 'Key', 'Range', 'Time_Signature', 'Tempo', 'Difficulty_Level'])
```

```python
i = 0
j = 0
for song in keylist:
    response = requests.get('https://www.flutetunes.com/tunes.php?id={}'.format(song))
    html_source = response.text
    soup = BeautifulSoup(html_source, 'html.parser')
#     df.loc[song, 'Title'] = re.findall("schema:name\">(.+?)</span",str(soup).decode('ascii', 'ignore'))[0]
    title = re.findall("schema:name\">(.+?)</span>",str(soup))[0]
    try:
        title2 = re.findall("schema:isPartOf\">(.+?)</span>",str(soup))[0]
        df.loc[song, 'Title'] = title + ' ' + title2
    except:
        df.loc[song, 'Title'] = title
        
    df.loc[song, 'Key'] = re.findall("schema:musicalKey\">(.+?)</td",str(soup))[0]
    df.loc[song, 'Time_Signature'] = re.findall("Time signature</th><td>(.+?)</td>",str(soup))[0]
    df.loc[song,'Range'] = re.findall("Range</th><td>(.+?)<sub>(.+?)</sub>\xe2\x80\x93(.+?)<sub>(.+?)</sub",str(soup))[0][0:4]
    df.loc[song, 'Tempo'] = re.findall("Tempo</th><td>(.+?)</td>", str(soup))[0]
    df.loc[song, 'Difficulty_Level'] = re.findall("Difficulty level</th><td style=\"padding-bottom: 1em;\">(.+?)</td>",str(soup))[0]
    try:
        df.loc[song, 'Link'] = re.findall('<a class="pdf" href="(.+?)" rel', str(soup))[0]
    except:
        df.loc[song, 'Link'] = 'NaN'
        print song
        j += 1
    i += 1
print j
    
```
